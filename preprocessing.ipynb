```python
import json
```

```python
with open('/content/Knowledge_가족과 주거_1.json', 'r') as f:

    data = json.load(f)
```

```python
data.keys()
```

```python
data
```

```python
context = data['utterances']
```

```python
context
```

```python
preprocessed_data = []

for i in range(len(context)-1):

    query = context[i]
    answer = context[i+1]

    d = {}

    knowledge = []

    if answer['reference_text'] is None:

        knowledge.append('')

    else:

        for v in answer['reference_text']:

            knowledge.append(v['value'])

    d['knowledge'] = knowledge
    d['query'] = query['text']
    d['answer'] = answer['text']

    preprocessed_data.append(d)
```
